```python
def bvplot(df,column,title,pad=3,r=45,w=20,h=25):
    sns.color_palette('dark')
    temp = df[column].value_counts()
    # Calculate the percentage of payment difficulties (target=1) in each category
    perc = df[[column, 'TARGET']].groupby([column],as_index=False).mean()
    perc.sort_values(by='TARGET', ascending=False, inplace=True)
    fig, axes = plt.subplots(2, 1, figsize=(w,h))
    fig.suptitle(title)
    fig.tight_layout(pad=pad)
    axes[0].tick_params(axis='x', rotation=r)
    axes[1].tick_params(axis='x', rotation=r)
    axes[0].set_title("Overall", y=0.8, x=0.8)
    axes[1].set_title("Defaulters Percentage", y=0.8, x=0.8)
    sns.barplot(ax = axes[0], x=temp.index, y=temp.values, palette='bright')
    sns.barplot(ax= axes[1], x=perc[column].to_list(), y=perc['TARGET'].to_list(), palette='magma')
    fig.show()
```
